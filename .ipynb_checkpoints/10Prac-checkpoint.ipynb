```python
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None
        self.level = None
```

```python
class Tree:
    def createNode(self,value):
        return Node(value)
    
    def insertNode(self,root,value):
        if root is None:
            return self.createNode(value)
        if root.value > value:
            root.left = self.insertNode(root.left,value)
        else:
            root.right = self.insertNode(root.right,value)
        return root
    
    def inOrderTraversal(self,root):
        if root is None:
            return
        self.inOrderTraversal(root.left)
        print(root.value)
        self.inOrderTraversal(root.right)
    
    def levelOrderTraversal(self,root):
        if root is None:
            return
        q = []
        q.append(root)
        while len(q):
            root = q.pop(0)
            print(root.value,end=" ")
            if root.left:
                q.append(root.left)
            if root.right:
                q.append(root.right)
        return
    
    def topView(self,root):
        hashMap = {}
        root.level = 0
        q = []
        q.append(root)
        while len(q):
            currNode = q.pop(0)
            if currNode.level not in hashMap:
                hashMap[currNode.level] = currNode.value
            if currNode.left:
                currNode.left.level = currNode.level - 1
                q.append(currNode.left)
            if currNode.right:
                currNode.right.level = currNode.level + 1
                q.append(currNode.right)
        for key in sorted(hashMap):
            print(hashMap[key])
    
    def isBST(self,root):
        arr = []
        def inOrderhelper(root):
            if root is None:
                return
            inOrderhelper(root.left)
            arr.append(root.value)
            inOrderhelper(root.right)
        inOrderhelper(root)
        
        for val in range(len(arr)-1):
            if not arr[val] < arr[val + 1]:
                return False
        return True
    
    def isBST(self,root):
        leftVal, rightVal = float("-inf"), float("inf")
        
        def isBSThelper(root,leftVal,rightVal):
            if root is None:
                return True
            
            if root.left and root.left.value >= root.value:
                return False
            if root.right and root.right.value <= root.value:
                return False
            if root.value >= rightVal or root.value <= leftVal:
                return False
            
            return isBSThelper(root.left,leftVal,root.value) and isBSThelper(root.right,root.value,rightVal)
        
        return isBSThelper(root,leftVal,rightVal)
    
    def isBalanced(self,root):
        
        def helper(root):
            if root is None:
                return 0
            lh = 1 + helper(root.left)
            if lh == -1:
                return -1
            rh = 1 + helper(root.right)
            if rh == -1:
                return -1
            if abs(lh-rh) > 1:
                return -1
            
            return max(lh,rh) + 1
        
        return helper(root) != -1
    
    def diameterBST(self,root):
        # 1) Longest path between two nodes
        # 2) Not neccesarily pass via root node
        
        def diameterBSThelper(root):
            if root is None:
                return [0,0]
            lh = diameterBSThelper(root.left)
            rh = diameterBSThelper(root.right)
            return [1+max(lh[0],rh[0]),max(lh[0]+rh[0]+2,lh[1],rh[1])]
        
        return diameterBSThelper(root)[1]
    
    def maxPathSum(self,root):
        
        def maxPathSumHelper(root):
            if root is None:
                return [0,float("-inf")]
            
            leftSum = maxPathSumHelper(root.left)
            rightSum = maxPathSumHelper(root.right)
            
            leftreturn = max(leftSum[0] + root.value, root.value, rightSum[0] + root.value)
            rightreturn = max(leftSum[0] + root.value + rightSum[0], leftSum[1], rightSum[1])
            
            return [leftreturn,rightreturn]
        
        return maxPathSumHelper(root)[1]
    
    def maxPath(self,root):
        def maxPathSumHelper(root):
            if root is None:
                return [0,float("-inf")]
            leftSum = maxPathSumHelper(root.left)
            rightSum = maxPathSumHelper(root.right)
            
            leftreturn = max(root.value,root.value+max(leftSum[0],rightSum[0]))
            rightreturn = max(leftSum[1], rightSum[1], root.value, root.value + leftSum[0] + rightSum[0])
            
        
        return maxPathSumHelper[1]
            
        
        
```

```python
tree = Tree()
root = tree.createNode(5)
```

```python
ele = "10 7 8 12 30 45"
elements = list(int(x) for x in ele.split(" "))

for element in elements:
    root = tree.insertNode(root,element)
    
```
