```python
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None
```

```python
class Tree:
    def nodesAtK(self,root,valNode,distance):
        parentPair = {}
        q = []
        q.append(root)
        while len(q):
            for i in range(len(q)):
                node = q.pop(0)
                if node.left:
                    parentPair[node.left] = node
                    q.append(node.left)
                if node.right:
                    parentPair[node.right] = node
                    q.append(node.right)
        queryQ = []
        queryQ.append(valNode)
        visitedNodes = []
        travelled = 0
        while travelled <= distance:
            for i in range(len(queryQ)):
                node = queryQ.pop(0)
                if node.left and node.left not in visitedNodes:
                    queryQ.append(node.left)
                if node.right and node.right not in visitedNodes:
                    queryQ.append(node.right)
                if node in parentPair and parentPair[node] not in visitedNodes:
                    queryQ.append(parentPair[node])
                visitedNodes.append(node)
            travelled += 1
        return queryQ
    
    def minBurnTime(self,root,valNode):
        q = []
        q.append(root)
        parentPair = {}
        while len(q):
            for i in range(len(q)):
                node = q.pop(0)
                if node.left:
                    q.append(node.left)
                    parentPair[node.left] = node
                if node.right:
                    q.append(node.right)
                    parentPair[node.right] = node
        visitedNodes = []
        timeTaken = 0
        q = []
        q.append(valNode)
        while len(q):
            flag = False
            for i in range(len(q)):
                node = q.pop(0)
                if node.left and node.left not in visitedNodes:
                    q.append(node.left)
                    flag = True
                if node.right and node.right not in visitedNodes:
                    q.append(node.right)
                    flag = True
                if node in parentPair and parentPair[node] not in visitedNodes:
                    q.append(parentPair[node])
                    flag = True
                visitedNodes.append(node)
            timeTaken += 1 if flag else 0
        return timeTaken
                    
        
```
