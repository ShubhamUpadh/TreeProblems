```python
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None
```

```python
class Tree:
    def createNode(self,value):
        return Node(value)
    
    def insertNode(Self,root,value):
        if root is None:
            return self.createNode(value)
        if root.value > value:
            root.left = self.insertNode(root.left,value)
        else:
            root.right = self.insertNode(root.right,value)
        return root
    
    def ZigZagTraversal(self,root):
        flag = True
        q = []
        q.append(root)
        retList = []
        while len(q):
            lenLevel = len(q)
            currLevel = []
            for i in lenLevel:
                root = q.pop(0)
                if root.left:
                    q.append(root.left)
                if root.right:
                    q.append(root.right)
                currLevel.append(root.value)
            if not flag:
                currLevel = currLevel[::-1]
            flag = not flag
            retList.append(currLevel)
        
        return retList           
    
    def BoundaryTraversal(self,root):
        #Left Boundary, leaf nodes, Right boundary in reverse
        retList = []
        rightBound = []
        def leftBoundary(node):
            if node is None:
                return
            if node.left.left == None and node.left.right == None and node.right.left == None and node.right.right == None:
                return node
            retList.append(node.value)
            if node.left:
                return leftBoundary(node.left)
            else:
                return leftBoundary(node.right)
        def AddLeaf(node):
            if node.left:
                retList.append(node.left.value)
            if node.right:
                retList.append(node.right.value)
            return
        def rightBoundary(node):
            if node is None:
                return
            if node.left.left == None and node.left.right == None and node.right.left == None and node.right.right == None:
                return node
            rightBound.append(node.value)
            if node.right:
                return leftBoundary(node.right)
            else:
                return leftBoundary(node.left)
        temproot = root
        leftLast = leftBoundary(temproot)
        AddLeaf(leftLast)
        temproot = root
        rightLast = rightBoundary(temproot)
        AddLeaf(rightLast)
        
        
        
```
